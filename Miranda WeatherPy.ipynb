#### Observable Trends
* Trend 1: The temperature peaks at a latitude of 20 degrees and decreases as the latitude diverges from 20 degrees. This is likely due to the position and tilt of the Earth, relative to the sun.
* Trend 2: Latitude does not appear to have a direct correlation with humidity. However, there does appear to be a greater number of cities with high % humidity than with lower % humidity.
* Trend 3: As with humidity, cloudiness does not appear to be diractly related to humidity. However, it does appear that most cities are most densely concentrated around very high or very low cloudiness levels. Cities are less densely concentrated around more moderate cloudiness levels.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

from pprint import pprint

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        
# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it's being processed (with the city number and city name).


```python
# Perform a weather check on each city using a series of successive API calls.

# City data to collect:
city_name = []
city_lat = []
city_lng = []
city_max_temp = []
city_humidity = []
city_cloudiness = []
city_wind = []
city_country = []
city_date = []

# Print to logger:
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Creating counters for printing log:
record_index = 1
set_index = 1

# Looping through all the cities and performing a request:
for index, city in enumerate(cities):

    url = "http://api.openweathermap.org/data/2.5/weather?"
    units = "imperial"
    
    # Build query URL
    query_url = url + "appid=" + weather_api_key + "&q=" + city + "&units=" + units

    # Check to see if the index is divisible by 50
    # If the index is divisible by 50, reset the current record
    # And increment the set number by 1
    if (record_index % 50 == 0 and record_index >= 50):
        record_index = 1
        set_index += 1
        
    # Run an API request for the city
    # Using try and except
    try:
        
        # Make request and store in json
        weather_response = requests.get(query_url).json()
        
        # Store city data in empty lists
        city_name.append(weather_response["name"])
        city_lat.append(weather_response["coord"]["lat"])
        city_lng.append(weather_response["coord"]["lon"])
        city_max_temp.append(weather_response["main"]["temp_max"])
        city_humidity.append(weather_response["main"]["humidity"])
        city_cloudiness.append(weather_response["clouds"]["all"])
        city_wind.append(weather_response["wind"]["speed"])
        city_country.append(weather_response["sys"]["country"])
        city_date.append(weather_response["dt"])
        
        # Print a log of each city that is processed
        print(f"Processing Record {record_index} of Set {set_index} | {weather_response['name']}")
        
        # Suspend time to prevent api key from getting locked
        time.sleep(2)
        
        # Increment the record index if the city was found
        record_index += 1
    
    # Handle exceptions if a city is not found
    except:
        print("City not found. Skipping...")

# Print a message when processing is complete
print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Create a city data dataframe
city_data_dictionary = {"City": city_name,
                        "Lat": city_lat,
                        "Lng": city_lng,
                        "Max Temp": city_max_temp,
                        "Humidity": city_humidity,
                        "Cloudiness": city_cloudiness,
                        "Wind Speed": city_wind,
                        "Country": city_country,
                        "Date": city_date}
city_data_df = pd.DataFrame(city_data_dictionary)
city_data_df.to_csv(r'/Users/mirandahermes/GT-VIRT-DATA-PT-05-2022-U-LOLC/Contents/test_output_data/city_data.csv', index = False)
city_data_df.head()
```

## Linear Regression

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Create a Lat v Temp Dataframe for Northern Hemisphere
north_lat_v_wind_df = clean_city_data.loc[clean_city_data['Lat']>=0,['Lat','Max Temp']]

# Create a linear regression line
# Identify the x and y values
x_values = north_lat_v_wind_df['Lat']
y_values = north_lat_v_wind_df['Max Temp']

# Determine the values for the linear regression line
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept, 2))
# Similar to the y values for the liner regression line
regress_values = x_values * slope + intercept

# Create a Lat v Wind Speed Scatterplot
plt.scatter(x_values, y_values)

# Plot the Linear Regression Line
plt.plot(x_values, regress_values, "r-")

# Print the r-value
print(f"The r-value is: {rvalue**2}")

# Annotate the Lin. Reg. Line and Scatter Plot
plt.annotate(line_eq,(10,35),fontsize=15,color="red")
plt.ylabel('Max Temp')
plt.xlabel('Latitude')
plt.show()
```
